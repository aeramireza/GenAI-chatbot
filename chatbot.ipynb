Let's establish a few common financial queries to extract information from the dataset:

1. What is Company X's total revenue for year Y?
2. What is Company X's net income for year Y?
3. How has Company X's total revenue changed over the last year?
4. What is Company X's equity to asset ratio for year Y?
5. What is Company X's cash flow from operating activities to net income ratio for year Y?



```python
def simple_chatbot(df):
    # This chatbot will identify the companies, financial figures and years that the user is interested in
    print("Bot: Hello! I'm a chatbot specialized in financial queries. How can I assist you today?")
    user_query = input("You: ").lower()

    # Initialize the variables the user is interested in and getting them from the dataset:
    company = None
    figure = None
    year = None

    companies = df['company'].unique()
    figures = df.columns[2:]
    years = df['year'].unique()

    for comp in companies:
        if comp.lower() in user_query:
            company = comp
            break

    for fig in figures:
        if fig.lower() in user_query:
            figure = fig
            break

    for num in years:
        if str(num) in user_query:
            year = num
            break

    if not company:
        company = input("Bot: Sorry, I couldn't identify the company in your query. Please specify it: ").capitalize()
    if not figure:
        figure = input("Bot: Sorry, I couldn't identify the financial figure in your query. Please specify it: ").split()
        figure = list(map(str.capitalize, figure))
        figure = ' '.join(figure)
    if not year:
        year = int(input("Bot: Sorry, I couldn't identify the year in your query. Please specify it: "))

    fetch = df[(df['company']==company) & (df['year']==year)][figure]

    if np.isnan(fetch.values[0]):
        print("Bot: Sorry, the information you are looking for cannot be obtained from the dataset provided.")
    elif not fetch.empty:
        if re.search(r"%", user_query):
            print("Bot: {}'s {} in {} was {}%.".format(company, re.sub(r"\s\(%\)$", "", figure), year, round(fetch.values[0], 1)))
        else:
            print("Bot: {}'s {} in {} was {} million USD.".format(company, figure, year, fetch.values[0]))
    else:
        print("Bot: Sorry, the information you are looking for cannot be obtained from the dataset provided.")
```

```python
simple_chatbot(df)
```
